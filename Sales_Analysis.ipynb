3. Data Cleaning
Data Cleaning Process

Data cleaning is one of the most important steps in any data analysis project because inaccurate or incomplete data can lead to misleading conclusions.

Several preprocessing techniques were applied to improve the quality of the dataset.

Handling Missing Values

The dataset was examined for missing values using the isnull().sum() function. Missing values were handled by either replacing them with appropriate values (such as zero or the column mean) or removing records where necessary.

This ensured that incomplete data would not affect the accuracy of the analysis.

Removing Duplicate Records

Duplicate rows were identified using the duplicated() function and removed using drop_duplicates().

Removing duplicate records prevents double-counting of transactions and improves data reliability.

Correcting Data Types

The Order Date column was converted to the DateTime format using pd.to_datetime(), allowing date-based analysis such as monthly and yearly sales trends.

Numeric columns such as Sales and Profit were also converted to the appropriate numeric data types.

Feature Engineering

Several new columns were created to improve the analysis.

These include:

Year: Extracted from the Order Date.
Month: Month name extracted from the Order Date.
Quarter: Financial quarter of each transaction.
Profit Margin: Calculated using the formula (Profit / Sales) × 100.
Sales Category: Categorized sales into Low, Medium, High, and Very High using predefined ranges.

These engineered features enabled more detailed business analysis and visualization.

4. Exploratory Data Analysis (EDA)
Exploratory Data Analysis

Exploratory Data Analysis (EDA) was performed to understand the dataset and identify meaningful business trends.

The following key performance indicators (KPIs) were calculated:

Total Revenue

The total revenue generated by the company was calculated by summing all sales transactions. This metric provides an overall view of business performance.

Total Profit

The total profit earned from all customer orders was calculated to evaluate the company's profitability.

Average Sales

Average sales per transaction were calculated to understand the typical value of customer purchases.

Average Profit

The average profit earned per order provides insight into the profitability of individual transactions.

Top Customers

Customer-wise sales analysis identified the highest-value customers contributing the largest share of total revenue.

Best-Selling Products

Products were ranked according to total sales, allowing identification of the company's best-performing products.

Category Performance

Sales were grouped by product category to determine which categories generated the highest revenue.

Regional Performance

Sales by region were analyzed to identify the strongest and weakest performing geographic markets.

Monthly Sales Trend

Monthly sales totals were calculated to identify seasonal patterns and peak business periods.

Overall, the EDA process transformed raw transactional data into meaningful business information that supports informed decision-making.

```python
df["ShippingCost"] = df["ShippingCost"].fillna(df["ShippingCost"].mean())
df["Category"] = df["Category"].fillna("unknown")
```

```python
df["OrderDate"] = pd.to_datetime(df["OrderDate"])
#df["TotalAmount"] = df["TotalAmount"].dtype(float)
```

```python
df["Year"] = df["OrderDate"].dt.year
df["Month"] = df["OrderDate"].dt.month
df["Quarter"] = df["OrderDate"].dt.quarter
```

```python
df["Profitmargin"] = (df["TotalAmount"]-df["Tax"]-df["UnitPrice"]-df["ShippingCost"])/df["TotalAmount"]
```

```python
df["SalesCategory"] = pd.cut(
    df["TotalAmount"],
    bins = [0, 500, 1000, 2000, 10000],
    labels= ["low", "average", "good", "high"]
)
```

```python
int(df["TotalAmount"].sum())
int(df["Profitmargin"].sum())
int(df["TotalAmount"].mean())
int(df["Profitmargin"].mean())
int(df["TotalAmount"].max())
int(df["TotalAmount"].min())



```

```python
df.groupby("CustomerName")["TotalAmount"].sum().sort_values(ascending=False).head(10)
```

5. Visualizations
Data Visualization

Visualizations were created to present the analysis in an easy-to-understand graphical format. Charts help identify trends, comparisons, and relationships that may not be immediately visible in raw numerical data.

Sales Trend

A line chart was created to display sales performance over time. This visualization highlights periods of high and low sales and helps identify seasonal demand patterns.

Category Comparison

A bar chart compares total sales across different product categories, making it easy to identify the highest-performing and lowest-performing categories.

Profit Distribution

A histogram illustrates the distribution of profit values across all transactions. This chart helps identify the frequency of profitable and loss-making orders.

Correlation Matrix

A correlation heatmap was generated using Plotly to examine relationships between numerical variables such as Sales, Profit, Quantity, and Discount.

This visualization helps determine whether increases in one variable are associated with increases or decreases in another.

Monthly Sales

A monthly line chart illustrates changes in sales throughout the year, enabling the identification of seasonal trends and peak business months.

Top 10 Products

A horizontal bar chart displays the ten highest-selling products based on total revenue.

This visualization helps management identify products that should receive greater marketing and inventory investment.

Regional Sales

A pie chart illustrates the percentage contribution of each region to total company sales, providing a clear understanding of regional market performance.

```python
df["date"] = df["OrderDate"].dt.to_period("M")
sales = df.groupby("date")["TotalAmount"].sum()
plt.figure(figsize=(12,6))
sales.plot()
plt.title("Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.savefig("./sales_trend.png")
plt.show()
```

```python
category = df.groupby("Category")["Profitmargin"].sum().sort_values(ascending=False).head(10)
category.plot(kind="bar", label= category.values)
plt.title("Sales by Category")
plt.savefig("./category.png")
plt.show()
```

```python

df["Profitmargin"].plot(kind="hist", bins = 20)
plt.title("Profit Distribution")
plt.savefig("./profit_hist.png")
plt.show()
```

```python
# 1. Calculate the correlation matrix
correlation = df.corr(numeric_only=True)

# 2. Create the heatmap with a larger width and height (e.g., 800x800 or 1000x1000 pixels)
fig = px.imshow(
    correlation, 
    text_auto=True, 
    title="Correlation Matrix",
    width=800,   # Set width in pixels
    height=800   # Set height in pixels
)

# 3. Optional: Adjust font size of the text annotations inside the squares
fig.update_layout(
    title_font_size=20,
    xaxis_tickfont_size=12,
    yaxis_tickfont_size=12
)

fig.show()
```

```python
product = product.sort_values(ascending=False).head(10)
product.plot(kind="barh")
plt.title("Top Products")
plt.savefig("./top_products.png")
plt.show()
```

```python
plt.figure(figsize=(8,8))
state.plot(kind="pie", labels=state.index, autopct="%1.1f%%")
plt.title("Regional Sales")
plt.savefig("./region.png")
plt.show()
```
